# IRIS DATASET

```python
import seaborn as sns 
df = sns.load_dataset('iris')
```

```python
df.head()
```

We want features which will help me in predicting the species of flower
and therefore will use pdf here

```python
sns.kdeplot(data=df,x='sepal_length',hue='species')
```

What do you infer from this plot?
The likeliness of flower being Setosa is if and only if sepal_length falls somewhat in the range (4.3-5.4) most probable region
(3.8-4.3) is also possible but lesser chance same with the range (5.4-6.3)

For versicolor most probable region is (5.3-6.5)
(4.3-5.3) is also possible but lesser chance same with the range (6.5-7.7)

For virginica most popular region is (5.7-7.8)
(4.2-5.7) is also possible but lesser chance same with the range (7.8-8.8)

Now looking at the most probable regions of all the three flower we can say that sepal length is less important to us since they are overlapping 

```python
sns.kdeplot(data=df,x='sepal_width',hue='species')
```

Unlike sepal_length we can say that all the three species are overlapping and hence less important 

```python
sns.kdeplot(data=df,x='petal_length',hue='species')
```

Here For setosa being setosa only needs petal length ranging from (1-1.9) we can confidently say that if the petal length is of this range the species is setosa 

similarly for versicolor, if the petal length ranges from    (3.5-4.8) then it is of versicolor species 

and for virginica the range is (4.7-6.4), which is also clearly differentiable 

And hence we can say that petal length is an important feature to condider

```python
sns.kdeplot(data=df,x='petal_width',hue='species')
```

Similary here also we can see that the species are differentiable and this feature can be considered as the important one 

For features sepal_length and sepal_width the machine learning model will find difficulty in understanding the differentiation as there is no clear separation 

Thus using this decision we can keep petal length and petal width as they show clear distinction
