# Standard Normal Distribution
Standard Normal Distribution is a normal distribution which has mean equal to 0 and std equal to 1

```python
import pandas as pd
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head(2)
```

```python
import seaborn as sns 
sns.kdeplot(df['Age'])
```

We can assume that this is normal distribution

```python
mean = df['Age'].mean()
```

```python
std = df['Age'].std()
```

```python
# X ~ N(29,14) where X is random variable (Age)

# goal is to convert it into standard normal deviation

# Z ~ N(0,1)
```

```python
snd_series = (df['Age']-mean)/std
```

```python
snd_series.mean()
```

```python
snd_series.std()
```

```python
# It wont be perfect zero since we didnt take perfect normal distribution
```

```python
sns.kdeplot(snd_series)
```
