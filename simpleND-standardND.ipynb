```python
import seaborn as sns 
sns.kdeplot(df['Age'])
```

We can assume that this is normal distribution

```python
mean = df['Age'].mean()
```

```python
std = df['Age'].std()
```

```python
# X ~ N(29,14) where X is random variable (Age)

# goal is to convert it into standard normal deviation

# Z ~ N(0,1)
```

```python
snd_series = (df['Age']-mean)/std
```

```python
snd_series.mean()
```

```python
snd_series.std()
```

```python
# It wont be perfect zero since we didnt take perfect normal distribution
```

```python
sns.kdeplot(snd_series)
```

Standard Normal Distribution is a normal distribution which has mean equal to 0 and std equal to 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 68
sigma = 3

# Create a range of x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the normal distribution
y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Plot the normal distribution
plt.plot(x, y, label='Normal Distribution')

# Highlight the region within ±3σ
x_fill = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y_fill = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_fill - mu) / sigma)**2)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label='Within ±3σ')

# Add labels and title
plt.title('Normal Distribution with ±3σ Region')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
