There are many ways of identifying whether the give data is normally distributed or not. The important ones includes 
1. Visual inspection
2. Statistical test : Anderson-Darling Test, Kolmogrovs-smirnov test 
3. QQ plot

# QQ Plot 
QQ plot is used to compare the similarity between the distribution of two sets of data
[X,Y] X is the given data and Y is the theoritical distribution lets say Normal distribution

Both sets are divided into quantiles lets say 100 and each percentile is plotted using scatter plot. If the plot stretches a straight line diagnally then X follows a Normal Distribution

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
```

```python
df = sns.load_dataset('iris')
```

```python
sns.kdeplot(df['sepal_length'])
```

u first sort the data then divide into quantiles and smae for the other and then plot the scatter plot 

```python
temp = sorted(df['sepal_length'].tolist())
# tolist() converts series into list 
```

```python
y_quant = []
for i in range(1,101):
    y_quant.append(np.percentile(temp,i))
```

```python
y_quant
```

```python
samples = np.random.normal(loc=0,scale=1,size=1000)
```

```python
x_quant = []
for i in range(1,101):
    x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

```python
# using statsmodel

import statsmodels.api as sm
import matplotlib.pyplot as plt 

# create a QQ plot of the two sets of data 
fig = sm.qqplot(df['sepal_length'],line='45',fit=True)

plt.title('QQ plot')
plt.xlabel('Theoretical Quantities')
plt.ylabel('Sample Quantiles')

plt.show()
```

```python
# Letss see a uniformly distributed data 

x = np.random.uniform(low=0, high=1, size=1000)
```

```python
plt.hist(x)
```

```python
# fit a uniform distribution  to the data 
import scipy.stats as stats
import statsmodels.api as sm 

params = stats.uniform.fit(x) # creating a theoritcal uniform distributed data so that we can compare it with the x data
dist = stats.uniform(loc=params[0],scale=params[1])
```

```python
# create a qq plot of the data using the uniform distribution 
fig = sm.qqplot(x, dist=dist, line='45')
# we used dist which is a parameter for uniform built externally 

# whereas in the case of normal distribution we need not to specify it 
plt.title('qqplot of uniform distribution with uniform fit')
plt.xlabel('Theoritical quantities')
plt.ylabel('sample Quantities')
```

So this plot shows us that x is a uniform distribution
